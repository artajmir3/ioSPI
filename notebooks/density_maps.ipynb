## Reading Density Maps
We can use `density_maps.read_density_map_from_mrc` function to read a density map from a `.mrc` file. 

```python
data = density_maps.read_density_map_from_mrc('data/EMD-5592.map')
```

## Visualization
For visualization purposes, we illustrate three side views of the given map.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,15))
ax1.imshow(data.sum(0))
ax2.imshow(data.sum(1))
ax3.imshow(data.sum(2))
plt.show()
```
